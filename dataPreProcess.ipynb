```python
# 处理解压失败的少量文件
import os

# 设置路径和年份范围
asc_start_year = 2022
asc_end_year = 2024
des_start_year = 2002
des_end_year = 2024

# Asc
asc_unzip_path = r"G:\data\VOD\unzip\MCCA-VOD(KuCX)\Asc_unzip" 
asc_rar_path = r"G:\data\VOD\mat\kuxcVOD\Zip_Asc"

# Des
des_unzip_path = r"G:\data\VOD\unzip\MCCA-VOD(KuCX)\Des_unzip"
des_rar_path = r"G:\data\VOD\mat\kuxcVOD\Zip_Des"

def check_unzip_status(rar_path, unzip_path, start_year, end_year, prefix):
    failed_files = []

    for year in range(start_year, end_year + 1):
        rar_file = os.path.join(rar_path, f"{prefix}_{year}.rar")
        out_dir = os.path.join(unzip_path, str(year))

        if not os.path.isfile(rar_file):
            print(f"压缩文件不存在: {rar_file}")
            failed_files.append(rar_file)
            continue

        # 如果解压目录不存在或者为空，认为解压失败
        if not os.path.exists(out_dir) or len(os.listdir(out_dir)) == 0:
            failed_files.append(rar_file)

    return failed_files

# 检查 Asc
asc_failed = check_unzip_status(asc_rar_path, asc_unzip_path, asc_start_year, asc_end_year, "Asc")
# 检查 Des
des_failed = check_unzip_status(des_rar_path, des_unzip_path, des_start_year, des_end_year, "Des")

# 输出结果
if asc_failed:
    print("Asc 解压失败的文件：")
    for f in asc_failed:
        print(f)
else:
    print("Asc 全部解压成功！")

if des_failed:
    print("Des 解压失败的文件：")
    for f in des_failed:
        print(f)
else:
    print("Des 全部解压成功！")
```
