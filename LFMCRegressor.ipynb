#### 1.1.1 根据Reference+Country，整理LFMC汇总的所有数据集以及每个数据集的条数

```python
"""
本代码实现以下功能：
1. 从指定路径 G:\data\Globe LFMC\Globe-LFMC-2.0 final.xlsx 中读取名为 "LFMC data" 的工作表。
2. 读取 Reference 列，并根据 Reference 的唯一内容进行分类统计，得到每个 Reference 对应的数据条数。
3. 将上述统计结果写入 G:\data\Globe LFMC\Reference.txt。
4. 读取 Country 列，与 Reference 结合进行双变量分类统计（即 Country + Reference）。
5. 将 Country、Reference 及其对应的数据行数整理为一个包含三列的表格：
       - Country
       - Reference
       - Count（该 Country+Reference 类别的总行数）
6. 将该统计表写入 G:\data\Globe LFMC\regionAndRef.xlsx。
"""

import pandas as pd
import os

# 1. 文件路径
excel_path = r"G:\data\Globe LFMC\Globe-LFMC-2.0 final.xlsx"
txt_output_path = r"G:\data\Globe LFMC\Reference.txt"
xlsx_output_path = r"G:\data\Globe LFMC\regionAndRef.xlsx"

# 2. 读取 Excel
df = pd.read_excel(excel_path, sheet_name="LFMC data")

# ==============================
# Part 1：Reference 分类统计并输出 txt
# ==============================

reference_counts = df["Reference"].value_counts()

# 写入 txt
with open(txt_output_path, "w", encoding="utf-8") as f:
    f.write("Reference 分类统计结果：\n\n")
    for ref, count in reference_counts.items():
        f.write(f"{ref}: {count} 条数据\n")

print("Reference 分类统计已保存至：", txt_output_path)

# ==============================
# Part 2：Country + Reference 分类统计并输出 Excel
# ==============================

# 按 Country 和 Reference 分类统计
group_counts = (
    df.groupby(["Country", "Reference"])
      .size()
      .reset_index(name="Count")   # 生成 Count 列
)

# 写入 Excel
group_counts.to_excel(xlsx_output_path, index=False)

print("Country + Reference 分类统计已保存至：", xlsx_output_path)
```
